```python
import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
```

```python
pd.set_option('max_rows', 20)

plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 3)
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# 指定檔案位置，包含完整路徑以及檔案名稱
# os.curdir: 回傳目前目錄的路徑
Turbofan_FILEPATH = os.path.join(os.curdir, 'train_PM.txt')

# 檢視檔案位置
print(Turbofan_FILEPATH)
```

```python
metadata = ['unit_number', 'time_in_cycles']
operational_settings = ['op_setting_{}'.format(i + 1) for i in range (3)]
sensor = ['sensor_{}'.format(i + 1) for i in range(27)]
col = metadata + operational_settings + sensor
```

```python
# 讀取 CSV 檔案，並將其指定為名稱是 df (DataFrame) 的物件
df = pd.read_csv(filepath_or_buffer=Turbofan_FILEPATH, sep=' ', header=None, names=col)
df_bak = df.copy()

# 使用 head() method 大略地瀏覽資料
df.head()
```

```python
df.to_csv(os.path.join(os.curdir, 'data', 'turbofan.csv'), index=False)
```
