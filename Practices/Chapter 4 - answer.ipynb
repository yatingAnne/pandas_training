# Chapter 4 - Filtering Data

## Q1. 請基於以下條件觀察鐵達尼號乘客資料的敘述性統計分析結果

條件：存活與否 (`Survived`) 資料中，記錄為「**存活**」的資料

> 備註：存活與否 (`Survived`) 資料中，`0` = 未存活，`1` = 存活

```python
is_survived = df['Survived'] == 1

df.loc[is_survived].describe()
```

## Q2. 請基於以下條件觀察鐵達尼號乘客資料的敘述性統計分析結果

1. 存活與否 (`Survived`) 資料中，記錄為「**存活**」
1. 年齡 (`Age`) 資料中，介於 `0` ~ `30` 歲（含上下界）

> 備註：存活與否 (`Survived`) 資料中，`0` = 未存活，`1` = 存活

```python
is_survived_age = (df['Survived'] == 1) & (df['Age'] >= 0) & (df['Age'] <= 30)

df.loc[is_survived_age].describe(include='all')
```

## Q3. 請基於以下條件觀察「年齡（`Age`）」資料的敘述性統計分析的結果

條件：排除掉年齡 (`Age`) 資料的遺漏值

```python
age_notna = df.loc[:, 'Age'].notna()

df.loc[age_notna, 'Age'].describe()
```
