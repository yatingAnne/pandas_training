# Chapter 6 - Modifying DataFrame

## Q1

請將「性別 (`Sex`)」資料中的 `male` 值取代為 `0`，`female` 值取代為 `1`，並取出前五筆資料確認結果是否正確：

```python
df['Sex'].replace({'male':0, 'female':1}, inplace=True)
df.head()
```

## Q3

將「船艙代號 (`Cabin`)」資料的遺漏值補入數值 `0`，並檢視此欄位不重複的所有資料：

```python
# 還原資料
df = df_bak.copy()
```

```python
df['Cabin'] = df['Cabin'].fillna(value=0)
df['Cabin'].unique()
```

## Q4

## Q5

請觀察相同的「性別 (`Sex`)」與「存活與否 (`Survived`)」條件的乘客資料中，「年齡」(`Age`)資料的敘述性統計分析結果：

```python
# 還原資料
df = df_bak.copy()
```

```python
df.groupby(['Sex', 'Survived'])['Age'].describe()
```

## 補充——畫出 MultiIndex 的 boxplot

課程中有介紹「單一 index 與 column」的 boxplot，但是像 Q5 為 MultiIndex 的情況，也可以畫出 boxplot：

```python
# 還原資料
df = df_bak.copy()
```

```python
boxplot = df.boxplot(column=['Age'], by=['Sex', 'Survived'], vert=False)
boxplot.set_title('')
```
